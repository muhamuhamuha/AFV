## DataFraming Totals Sheets

```python
def find_end(sheet_obj, row):
    cell = '$A$' + str(row)
    all_cells = []
    while True:
        if cell == '$XFD$' + str(row):
            break
        all_cells.append(cell)
        cell = sheet_obj[cell].end('right').address
    return all_cells[-1]
```
